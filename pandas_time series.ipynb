##시간 범위의 주기 설정하기

data_range()내에 있는 freq속성 이용하기

```python
#data_range()메서드의 freq인잣값을 B로 설정
#freq='B' 이면 평일만 포함시킨 DatatimeIndex 를 만듦.
print(pd.date_range('2017-01-01','2017-01-07',freq="B"))
```

###9.1.1 K-평균 

**k-평균 알고리즘**

**<최적의 클러스터 개수 찾기>**

**실루엣 점수(silhouette score)**  
(b-a)/max(a,b)  
a: 동일한 클러스터에 있는 다른 샘플까지 평균 거리(즉, 클러스터 내부의 평균 거리)  
b : 가장 가까운 클러스터까지 평균 거리(즉, 가장 가까운 클러스터의 샘플까지 평균 거리)  
-1 < silhouettte score < 1  
+1 에 가까우면 자신의 클러스터 안에 잘 속해 있고 다른 클러스터와는 멀리 떨어져 있다는 뜻이다.  
 0에 가까우면 클러스터 경계에 위치한다는 의미다.  
-1에 가까우면 이 샘플이 잘못된 클러스터에 할당되었다.  

**실루엣 다이어그램**  
높이 : 클러스터가 포함하고 있는 샘플의 개수  
너비 : 이 클러스터에 포함된 샘플의 정렬된 실루엣 계수  
수직 파선 : 각 클러스터 개수에 해당하는 실루엣 점수  
(한 클러스터의 샘플 대부분이 이 점수보다 낮은 계수를 가지면 클러스터의 샘플이 다른 클러스터랑 너무 가깝다는 것을 의미하므로 나쁜 클러스터이다)

###9.1.2 k-평균의 한계

1) 최적이 아닌 솔루션을 피하려면 알고리즘을 여러 번 실행 해야 함.  
2) k-평균은 클러스터의 크기나 밀집도가 서로 다르거나 원형이 아닐 경우 잘 작동하지 않는다.  

TIP: K-평균을 실행하기 전에 입력 특성의 스케일을 맞추는 것이 중요하다.  특성의 스케일을 맞추어도 모든 클러스터가 잘 구분되고 원형을 형태를 가진다고 보장할 수는 없지만 일반적으로 더 좋아진다.

##9.2 가우시안 혼합

샘플이 파라미터가 알려지지 않은 여러 개의 혼합된 가우시안 분포에서 생성되었다고 가정하는 확률 모델  
하나의 가우시안 분포에서 생성된 모든 샘플은 하나의 클러스터를 형성한다. 일반적으로 이 클러스터는 타원형이다. 

```python
from sklearn.datasets import make_moons
```

```python
X,y = make_moons(n_samples=1000,noise=0.05)
```

```python
from sklearn.mixture import GaussianMixture

gm = GaussianMixture(n_components=3, n_init=10)
gm.fit(X)
```

```python
gm.weights_
```

```python
gm.means_
```

```python
gm.covariances_
```

```python
#알고리즘 수렴 여부
gm.converged_
```

```python
#알고리즘 반복 횟수
gm.n_iter_
```

```python
#하드 군집
gm.predict(X)
```

```python
#소프트 군집
gm.predict_proba(X)
```

```python
X_new,y_new = gm.sample(6)
X_new
```

```python
y_new
```

```python
#주어진 위치에서 모델의 밀도를 측정
#샘플이 주어지면 이 메서드는 그 위치의 확률 밀도 함수(pdf)의 로그를 예측한다.
#점수가 높을수록 밀도가 높은 것
gm.score_samples(X)
```

확률 밀도: 0에서 1까지의 값이 아니라 어떤 양숫값도 될 수 있다.  
샘플이 특정 지역 안에 속할 확률을 예측하려면 그 지역에 대해 PDF를 적분해야 한다.

###9.2.1 가우시안 혼합을 사용한 이상치 탐지

밀도가 낮은 지역에 있는 모든 샘플을 이상치로 볼 수 있다.  
먼저 사용할 밀도 임곗값을 정해야 한다.

```python
densities = gm.score_samples(X)
```

```python
densities.shape
```

```python
import numpy as np
density_threshold = np.percentile(densities,4) #네 번째 백분위수(4%)를 밀도 임곗값으로 사용하여 이상치 구분
anomalies = X[densities < density_threshold]
```

```python
anomalies
```

###9.2.2 클러스터 개수 선택하기

가우시안 혼합에서는 이너셔나 실루엣 점수와 같은 지표를 사용할 수 없음. 이런 지표들은 클러스터가 타원형이거나 크기가 다를 때 안정적이지 않기 떄문이다.



*   BIC = log(m)p - 2log(L^)
*   AIC = 2p - 2log(L^)



m : 샘플의 개수  
p : 모델이 학습할 파라미터 개수  
L^ : 모델의 가능도 함수(likelihood function)의 최댓값

BIC와 AIC 는 모두 학습할 파라미터가 많은(즉 클러스터가 많은) 모델에게 벌칙을 가하고 데이터에 잘 학습하는 모델에게 보상을 더한다.

```python
gm.bic(X)
```

```python
gm.aic(X)
```

###9.2.3 베이즈 가우시안 혼합 모델

최적의 클러스터 개수를 수동으로 찾지 않고 불필요한 클러스터의 가중치를 0으로 (또는 0에 가깝게) 만드는 BayesianGaussianMixture 클래스를 사용할 수 있다.

```python
from sklearn.mixture import BayesianGaussianMixture
bgm = BayesianGaussianMixture(n_components=10,n_init=10)
bgm.fit(X)
np.round(bgm.weights_,2)
```
